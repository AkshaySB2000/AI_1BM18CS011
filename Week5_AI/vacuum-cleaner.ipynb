```python
def clean(floor):
    i, j, row, col = 0, 0, len(floor), len(floor[0])
    for i in range(row):
        if(i%2 == 0):
            for j in range(col):
                if(floor[i][j] == 1):
                    print_floor(floor, i, j)
                    floor[i][j] = 0
                print_floor(floor, i, j)
        else:
            for j in range(col-1, -1, -1):
                if(floor[i][j] == 1):
                    print_floor(floor, i, j)
                    floor[i][j] = 0
                print_floor(floor, i, j)
```

```python
def print_floor(floor, i, j): # row, col represent the current vacuum cleaner position
    print("The floor matrix is as below:")
    for r in range(len(floor)):
        for c in range(len(floor[r])):
            if r == i and c == j:
                 print(f" [{floor[r][c]}] ", end = '')
            else:
                print(f"  {floor[r][c]}  ", end = '')
        print(end = '\n')
    print(end = '\n')
```

```python
def main():
    print("Enter the number of rows")
    m = int(input())
    print("Enter the number of columns")
    n = int(input())
    floor = []
    i = 0
    print("Enter clean status for each cell (1 - dirty, 0 - clean)")
    for i in range(m):
        f = list(map(int, input().split(" ")))
        floor.append(f)
    print()
    clean(floor)
```
