```python
class Node:
    def __init__(self, data, level, fval):
        self.data = data
        self.level = level
        self.fval = fval

    def gen(self):
        a, b = self.find(self.data, '_')
        val_list = [[a, b - 1], [a, b + 1], [a - 1, b], [a + 1, b]]
        children = []
        for i in val_list:
            c = self.shuffle(self.data, a, b, i[0], i[1])
            if c is not None:
                child_node = Node(c, self.level + 1, 0)
                children.append(child_node)
        return children

    def shuffle(self, puz, a1, b1, a2, b2):
        if a2 >= 0 and a2 < len(self.data) and b2 >= 0 and b2 < len(self.data):
            tempp = []
            tempp = self.copy(puz)
            temp = tempp[a2][b2]
            tempp[a2][b2] = tempp[a1][b1]
            tempp[a1][b1] = temp
            return tempp
        else:
            return None

    def copy(self, r):
        temp = []
        for i in r:
            t = []
            for j in i:
                t.append(j)
            temp.append(t)
        return temp

    def find(self, puz, x):
        for i in range(0, len(self.data)):
            for j in range(0, len(self.data)):
                if puz[i][j] == x:
                    return i, j


class Puzzle:
    def __init__(self, size):
        self.n = size
        self.open = []
        self.closed = []

    def accept(self):
        puz = []
        for i in range(0, self.n):
            temp = input().split(" ")
            puz.append(temp)
        return puz

    def heur(self, start, target):
        return self.h(start.data, target) + start.level

    def h(self, start, target):
        temp = 0
        for i in range(0, self.n):
            for j in range(0, self.n):
                if start[i][j] != target[i][j] and start[i][j] != '_':
                    temp += 1
        return temp

    def process(self):
        print("Enter the start state matrix \n")
        start = self.accept()
        print("Enter the goal state matrix \n")
        target = self.accept()
        start = Node(start, 0, 0)
        start.fval = self.heur(start, target)
        self.open.append(start)
        print("\n\n")
        while True:
            curr = self.open[0]
            print("")
            for i in curr.data:
                for j in i:
                    print(j, end=" ")
                print("")
            print("\n")
            if (self.h(curr.data, target) == 0):
                break
            for i in curr.gen():
                i.fval = self.heur(i, target)
                self.open.append(i)
            self.closed.append(curr)
            del self.open[0]
            self.open.sort(key=lambda x: x.fval, reverse=False)
```

```python
puz = Puzzle(3)
puz.process()
```
